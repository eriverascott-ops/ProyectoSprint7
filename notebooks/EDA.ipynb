```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv("/content/drive/MyDrive/Triplete/Data Analyst/Modulo 2/vehicles_us.csv")

# Histograma agrupado por ao y marca

car_data['brand'] = car_data['model'].str.split().str[0]

fig = px.histogram(
    car_data,
    x="model_year",
    color="brand",
    barmode="group",
    title="Cantidad de carros por ao y marca"
)
fig.show()

# Grfico de dispersin odmetro vs precio
fig = px.scatter(
    car_data,
    x="odometer",
    y="price",
    color="model",
    title="Relacin entre Kilometraje y Precio",
    labels={"odometer": "Kilometraje (millas)", "price": "Precio en USD"}
)
fig.show()
```
